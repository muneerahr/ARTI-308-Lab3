## Age Distribution:

```python
plt.figure(figsize=(8,5))
plt.hist(df['RIDAGEYR'], bins=20)
plt.title("Distribution of Age (RIDAGEYR)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The plot shows that most individuals in the dataset are young. 
As age increases, the number of participants decreases.

## Bivariate Analysis

```python
df['RIAGENDR'] = df['RIAGENDR'].replace({1: "Male", 2: "Female"})

age_gender = df.groupby('RIAGENDR')['RIDAGEYR'].mean()
plt.figure(figsize=(6,4))
age_gender.plot(kind='bar')
plt.title("Average Age by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Age")
plt.show()
```

chart shows the average age for male and female participants.
Females have a slightly higher average age than males.


```python
bmi_gender = df.groupby('RIAGENDR')['RIDEXAGM'].mean()

plt.figure(figsize=(6,4))
bmi_gender.plot(kind='bar')
plt.title("Average BMI by Gender")
plt.xlabel("Gender")
plt.ylabel("Average BMI")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['RIDAGEYR'], df['DMDHHSIZ'], alpha=0.3)
plt.title("Age vs Household Size")
plt.xlabel("Age")
plt.ylabel("Household Size")
plt.show()
```

The scatter plot shows the relationship between age and household size. 
The points are distributed across all age groups without a clear increasing or decreasing pattern. 


```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(df[['RIDAGEYR','RIDEXAGM','DMDHHSIZ','INDFMPIR']].corr(), 
            annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
age_income = df.groupby('RIDAGEYR')['INDFMPIR'].mean()

plt.figure(figsize=(10,5))
age_income.plot()
plt.title("Income Ratio Trend by Age")
plt.ylabel("Average Income Ratio")
plt.xlabel("Age")
plt.show()
```

 The line plot illustrates the variation in average income ratio across different ages.
Income levels tend to increase during early adulthood and reach a peak in middle age.
After around age 60, a slight decline can be observed
